```python
column_names = ['X_input', 'y_prediction']

tribuo_lr_sgd_validation = pd.read_csv(f'{DATASETS_FOLDER}/tribuo-linear-regression-sgd-validation.csv')
tribuo_lr_sgd_validation.columns = column_names

tribuo_lr_ada_validation = pd.read_csv(f'{DATASETS_FOLDER}/tribuo-linear-regression-ada-validation.csv')
tribuo_lr_ada_validation.columns = column_names
tribuo_lr_cart_validation = pd.read_csv(f'{DATASETS_FOLDER}/tribuo-linear-regression-cart-validation.csv')
tribuo_lr_cart_validation.columns = column_names
tribuo_lr_xgb_validation = pd.read_csv(f'{DATASETS_FOLDER}/tribuo-linear-regression-xgb-validation.csv')
tribuo_lr_xgb_validation.columns = column_names
```

```python
def geometric_mean_of(distribution):
    index = 0
    N = len(distribution)
    result = 1
    for index in range(0, N):
        result = result * distribution[index]
    if N != 0:
        return abs(result) ** (1 / N)
    else:
        raise ValueError('No distribution has been passed in, cannot compute Geometric Mean')
```

```python
def harmonic_mean_of(distribution):
    index = 0
    N = len(distribution)
    result = 0
    for index in range(0, N):
        if distribution[index] != 0:
            result = result + (1 / distribution[index])

    if result == 0:
        raise ValueError('Distribution contains one or more zeros, cannot compute Harmonic Mean')
    else:
        return 1 / result
```

### Ensembling by averaging method (Tribuo Linear Regression: ADA, CART and XGBoost)

```python
tribuo_3_lr_ensembles_average = tribuo_lr_ada_validation.copy()
tribuo_3_lr_ensembles_average.y_prediction = \
           (tribuo_lr_ada_validation.y_prediction + tribuo_lr_cart_validation.y_prediction + \
            tribuo_lr_xgb_validation.y_prediction) / 3
tribuo_3_lr_ensembles_average.to_csv(f'{DATASETS_FOLDER}/tribuo-3-lr-ensembles-average.csv',index=False)
```

```python
plt.rcParams['figure.figsize'] = [20.0, 10.0]
tribuo_lr_ada_validation.y_prediction.plot(linestyle="dotted")
tribuo_lr_cart_validation.y_prediction.plot(linestyle="dashed")
tribuo_lr_xgb_validation.y_prediction.plot(linestyle="dashdot")
tribuo_3_lr_ensembles_average.y_prediction.plot(color="blue")
plt.legend(['tribuo: Linear Regression/ADA', 
            'tribuo: Linear Regression/CART', 
            'tribuo: Linear Regression/XGBoost',
            'tribuo: Linear Regression/Ensemble (average)'])
```

### Ensembling by median method (Tribuo Linear Regression: ADA, CART and XGBoost)

```python
tribuo_3_lr_ensembles_median = tribuo_lr_ada_validation.copy()
concatenated_df = pd.concat([tribuo_lr_ada_validation.y_prediction, 
                     tribuo_lr_cart_validation.y_prediction, 
                     tribuo_lr_xgb_validation.y_prediction], axis=1)
tribuo_3_lr_ensembles_median.y_prediction = np.median(concatenated_df, axis=1)
tribuo_3_lr_ensembles_median.to_csv(f'{DATASETS_FOLDER}/tribuo-3-lr-ensembles-median.csv',index=False)
```

```python
plt.rcParams['figure.figsize'] = [20.0, 10.0]
tribuo_lr_ada_validation.y_prediction.plot(linestyle="dotted")
tribuo_lr_cart_validation.y_prediction.plot(linestyle="dashed")
tribuo_lr_xgb_validation.y_prediction.plot(linestyle="dashdot")
tribuo_3_lr_ensembles_median.y_prediction.plot(color="blue")
plt.legend(['tribuo: Linear Regression/ADA', 
            'tribuo: Linear Regression/CART', 
            'tribuo: Linear Regression/XGBoost',
            'tribuo: Linear Regression/Ensemble (median)'])
```

### Ensembling by geometric mean method (Tribuo Linear Regression: ADA, CART and XGBoost)

```python
tribuo_3_lr_ensembles_geomean = tribuo_lr_ada_validation.copy()
tribuo_3_lr_ensembles_geomean.y_prediction = concatenated_df.y_prediction.apply(geometric_mean_of, axis=1)
tribuo_3_lr_ensembles_geomean.to_csv(f'{DATASETS_FOLDER}/tribuo-3-lr-ensembles-geomean.csv',index=False)
```

```python
plt.rcParams['figure.figsize'] = [20.0, 10.0]
tribuo_lr_ada_validation.y_prediction.plot(linestyle="dotted")
tribuo_lr_cart_validation.y_prediction.plot(linestyle="dashed")
tribuo_lr_xgb_validation.y_prediction.plot(linestyle="dashdot")
tribuo_3_lr_ensembles_geomean.y_prediction.plot(color="blue")
plt.legend(['tribuo: Linear Regression/ADA', 
            'tribuo: Linear Regression/CART', 
            'tribuo: Linear Regression/XGBoost',
            'tribuo: Linear Regression/Ensemble (geomean)'])
```

### Ensembling by harmonic mean method (Tribuo Linear Regression: ADA, CART and XGBoost)

```python
tribuo_3_lr_ensembles_harmonic_mean = tribuo_lr_ada_validation.copy()
tribuo_3_lr_ensembles_harmonic_mean.y_prediction = concatenated_df.y_prediction.apply(harmonic_mean_of, axis=1)
tribuo_3_lr_ensembles_harmonic_mean.to_csv(f'{DATASETS_FOLDER}/tribuo-3-lr-ensembles-harmonic-mean.csv',index=False)
```

```python
plt.rcParams['figure.figsize'] = [20.0, 10.0]
tribuo_lr_ada_validation.y_prediction.plot(linestyle="dotted")
tribuo_lr_cart_validation.y_prediction.plot(linestyle="dashed")
tribuo_lr_xgb_validation.y_prediction.plot(linestyle="dashdot")
tribuo_3_lr_ensembles_harmonic_mean.y_prediction.plot(color="blue")
plt.legend(['tribuo: Linear Regression/ADA', 
            'tribuo: Linear Regression/CART', 
            'tribuo: Linear Regression/XGBoost',
            'tribuo: Linear Regression/Ensemble (harmonic mean)'])
```
